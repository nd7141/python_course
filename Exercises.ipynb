```python
lst1 = ['a', 'b', 'c', 'd']
lst2 = ['b', 'c']
lst3 = ['c', 'd', 'e']
lst = [lst1, lst2, lst3]


d = {}
def map_f(list_l):
    d = {}
    for el in list_l:
        d[el] = d.get(el, 0) + 1
    return d

def reduce_f(acc, new_d):
    for (k,v) in new_d.iteritems():
        acc[k] = acc.get(k, 0) + v
    return acc


l_of_d = map(map_f, lst)
reduce(reduce_f, l_of_d)
```
